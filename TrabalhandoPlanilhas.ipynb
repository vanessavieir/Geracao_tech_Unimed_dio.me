```python
# Importando pacote de desenvolvimento
import pandas as pd
```

Leitura dos arquivos


```python
df1=pd.read_excel("/content/Aracaju.xlsx")
```

```python
df2=pd.read_excel("/content/Fortaleza.xlsx")
```

```python
df3=pd.read_excel("/content/Natal.xlsx")
```

```python
df4=pd.read_excel("/content/Recife.xlsx")
```

```python
df5=pd.read_excel("/content/Salvador.xlsx")
```

```python
# Visullizando as primeiras linhas
df1.head()
```

```python
# Concatenando os bancos de dados
df=pd.concat([df1,df2,df3,df4,df5])
```

```python
df.head()
```

```python
# visualizando as últimas linhas
df.tail()
```

```python
# Visualizando amostra dos dados
df.sample(5)
```

```python
# Tipoes de dados
df.dtypes
```

```python
# Transformando tipos inteiros em objeto
df["LojaID"]=df["LojaID"].astype("object")
```

```python
df.dtypes
```

```python
df.head()
```

```python
# Detectando se alguma linha possui um valor nulo
df.isnull().sum()
```

```python
# eliminando os valores nulos das linhas e colunas
df.dropna(inplace=True)
```

```python
# Apresentando a média da coluna vendas
df["Vendas"].mean()
```

```python
# Criando uma nova coluna
df["receita"]=df["Vendas"].mul(df["Qtde"])
```

```python
df.head()
```

```python
# valor máximo da receita
df["receita"].max()
```

```python
# valor mínimo da receita
df["receita"].min()
```

```python
# Apresentando as três maiores receitas
df.nlargest(3,"receita")
```

```python
# Apresentando 3 menores receitas
df.nsmallest(3,"receita")
```

```python
# Agrupando a soma das receitas por cidade
df.groupby("Cidade")["receita"].sum()
```

```python
df.sort_values("receita", ascending=False).head()
```

```python
# exemplo de como transformar data em inteiro
df["Data"]=df["Data"].astype("int64")
```

```python
df.dtypes
```

```python
# Transformando data, que está em inteiro , em data novamente
from pandas.io.sql import to_datetime
df["Data"]=to_datetime(df["Data"])
```

```python
df.dtypes
```

```python
# Agrupando o valor da receita por ano
df.groupby(df["Data"].dt.year)["receita"].sum()
```

```python
df.sample(5)
```

```python
# Aprentando anos de venda
df["ano_venda"]=df["Data"].dt.year
```

```python
df.sample(5)
```

```python
# meses e dias de venda
df["mes_venda"], df["dia_venda"]=(df["Data"].dt.month, df["Data"].dt.day)
```

```python
df.sample(5)
```

```python
# Valor mínimo de data
df["Data"].min()
```

```python
# Diferença entre as datas
df["diferenca_dias"]=df["Data"]- df["Data"].min()
```

```python
df.sample(5)
```

```python
# Apresentando os dados por trimestre
df["Trimestre"]=df["Data"].dt.quarter
```

```python
df.sample(5)
```

```python
# Apresentando dados por ano e mês pré-estabelecido
vendas_19_mar= df.loc[(df["Data"].dt.year==2019) & (df["Data"].dt.month==3)]
```

```python
vendas_19_mar
```

```python
# Verificando total de linhas com LojaID
df["LojaID"].value_counts(ascending=False)
```

```python
# Gráfico de barras com os valores acima
df["LojaID"].value_counts(ascending=False).plot.bar()
```

```python
# gráfico com barras horizontais, em ordem crescente
df["LojaID"].value_counts(ascending=True).plot.barh()
```

```python
# Criando um gráfico de pizza
df.groupby(df["Data"].dt.year)["receita"].sum().plot.pie()
```

```python
# Contagem de valores por cidade
df["Cidade"].value_counts()
```

```python
# Adicionando um título e alterando o nome dos eixos
import matplotlib.pyplot as plt
df["Cidade"].value_counts().plot.bar(title="Total vendas por cidade")
plt.xlabel("cidade")
plt.ylabel("Total vendas");
```

```python
# Alterando a cor do gráfico
df["Cidade"].value_counts().plot.bar(title="Total vendas por cidade", color="pink")
plt.xlabel("cidade")
plt.ylabel("Total vendas");
```
